```python
#modules
%matplotlib notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#create dataframe reference
df2018 = pd.read_csv('Project1CSV/Niceride 2010to2018/NR2018/data_2018.csv')
df2017 = pd.read_csv('Project1CSV/Niceride 2010to2018/NR2017/Nice_ride_trip_history_2017_season.csv')
df2016 = pd.read_csv('Project1CSV/Niceride 2010to2018/NR2016/Nice_ride_trip_history_2016_season.csv')
df2015 = pd.read_csv('Project1CSV/Niceride 2010to2018/NR2015/Nice_ride_trip_history_2015_season.csv')
df2014 = pd.read_csv('Project1CSV/Niceride 2010to2018/NR2014/Nice_ride_trip_history_2014_season.csv')
df2013 = pd.read_csv('Project1CSV/Niceride 2010to2018/NR2013/Nice_ride_trip_history_2013_season.csv')
df2012 = pd.read_csv('Project1CSV/Niceride 2010to2018/NR2012/Nice_ride_trip_history_2012_season.csv')
df2011 = pd.read_csv('Project1CSV/Niceride 2010to2018/NR2011/Nice_ride_trip_history_2011_season.csv')
df2010 = pd.read_csv('Project1CSV/Niceride 2010to2018/NR2010/Nice_ride_trip_history_2010_season.csv')
```
