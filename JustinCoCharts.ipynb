```python
#Count total of user types (cumulative)
usertype = df2018["usertype"].value_counts()
labels = ["Member", "Casual"]
```

```python
#Create pie chart showing overall (ALL YEARS) cumulative totals

plt.pie(usertype, labels = labels, autopct='%1.1f%%', shadow=True, startangle=140) 
plt.title("Overall User Type Count\n"+"(Cumulative)")
plt.show()
```
