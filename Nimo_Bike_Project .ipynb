```python
#DataFrames
nice_ride_2010_df = pd.read_csv(nice_ride_2010, low_memory=False).dropna(how="any")
nice_ride_2011_df = pd.read_csv(nice_ride_2011, low_memory=False).dropna(how="any")
nice_ride_2012_df = pd.read_csv(nice_ride_2012, low_memory=False).dropna(how="any")
nice_ride_2013_df = pd.read_csv(nice_ride_2013, low_memory=False).dropna(how="any")
nice_ride_2014_df = pd.read_csv(nice_ride_2014, low_memory=False).dropna(how="any")
nice_ride_2015_df = pd.read_csv(nice_ride_2015, low_memory=False).dropna(how="any")
nice_ride_2016_df = pd.read_csv(nice_ride_2016, low_memory=False).dropna(how="any")
nice_ride_2017_df = pd.read_csv(nice_ride_2017, low_memory=False).dropna(how="any")
nice_ride_2018_df = pd.read_csv(nice_ride_2018, low_memory=False).dropna(how="any")
```

```python
#Adding month column to data (2010)
nice_ride_2010_df["Month Number"] = pd.DatetimeIndex(nice_ride_2010_df["Start date"]).month
nice_ride_2010_df["Month Name"] = nice_ride_2010_df["Month Number"].apply(lambda x: calendar.month_name[x])

#Adding month column to data (2011)
nice_ride_2011_df["Month Number"] = pd.DatetimeIndex(nice_ride_2011_df["Start date"]).month
nice_ride_2011_df["Month Name"] = nice_ride_2011_df["Month Number"].apply(lambda x: calendar.month_name[x])

#Adding month column to data (2012)
nice_ride_2012_df["Month Number"] = pd.DatetimeIndex(nice_ride_2012_df["Start date"]).month
nice_ride_2012_df["Month Name"] = nice_ride_2012_df["Month Number"].apply(lambda x: calendar.month_name[x])

#Adding month column to data (2013)
nice_ride_2013_df["Month Number"] = pd.DatetimeIndex(nice_ride_2013_df["Start date"]).month
nice_ride_2013_df["Month Name"] = nice_ride_2013_df["Month Number"].apply(lambda x: calendar.month_name[x])

#Adding month column to data (2014)
nice_ride_2014_df["Month Number"] = pd.DatetimeIndex(nice_ride_2014_df["Start date"]).month
nice_ride_2014_df["Month Name"] = nice_ride_2014_df["Month Number"].apply(lambda x: calendar.month_name[x])

#Adding month column to data (2015)
nice_ride_2015_df["Month Number"] = pd.DatetimeIndex(nice_ride_2015_df["Start date"]).month
nice_ride_2015_df["Month Name"] = nice_ride_2015_df["Month Number"].apply(lambda x: calendar.month_name[x])

#Adding month column to data (2016)
nice_ride_2016_df["Month Number"] = pd.DatetimeIndex(nice_ride_2016_df["Start date"]).month
nice_ride_2016_df["Month Name"] = nice_ride_2016_df["Month Number"].apply(lambda x: calendar.month_name[x])

#Adding month column to data (2017)
nice_ride_2017_df["Month Number"] = pd.DatetimeIndex(nice_ride_2017_df["Start date"]).month
nice_ride_2017_df["Month Name"] = nice_ride_2017_df["Month Number"].apply(lambda x: calendar.month_name[x])

```

```python
months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
#calendar.month_name[months.index(monthname)+1]
nice_ride_2010_df["Month Name"] = pd.Categorical(nice_ride_2010_df["Month Name"], categories=months, ordered=True)
nice_ride_2011_df["Month Name"] = pd.Categorical(nice_ride_2011_df["Month Name"], categories=months, ordered=True)
nice_ride_2012_df["Month Name"] = pd.Categorical(nice_ride_2012_df["Month Name"], categories=months, ordered=True)
nice_ride_2013_df["Month Name"] = pd.Categorical(nice_ride_2013_df["Month Name"], categories=months, ordered=True)
nice_ride_2014_df["Month Name"] = pd.Categorical(nice_ride_2014_df["Month Name"], categories=months, ordered=True)
nice_ride_2015_df["Month Name"] = pd.Categorical(nice_ride_2015_df["Month Name"], categories=months, ordered=True)
nice_ride_2016_df["Month Name"] = pd.Categorical(nice_ride_2016_df["Month Name"], categories=months, ordered=True)
nice_ride_2017_df["Month Name"] = pd.Categorical(nice_ride_2017_df["Month Name"], categories=months, ordered=True)
```

```python
nice_ride_2010_df = nice_ride_2010_df.sort_values(by=["Month Number"], ascending=True)
nice_ride_2011_df = nice_ride_2011_df.sort_values(by=["Month Number"], ascending=True)
nice_ride_2012_df = nice_ride_2012_df.sort_values(by=["Month Number"], ascending=True)
nice_ride_2013_df = nice_ride_2013_df.sort_values(by=["Month Number"], ascending=True)
nice_ride_2014_df = nice_ride_2014_df.sort_values(by=["Month Number"], ascending=True)
nice_ride_2015_df = nice_ride_2015_df.sort_values(by=["Month Number"], ascending=True)
nice_ride_2016_df = nice_ride_2016_df.sort_values(by=["Month Number"], ascending=True)
nice_ride_2017_df = nice_ride_2017_df.sort_values(by=["Month Number"], ascending=True)
```

```python
#Counting Months
month_count_2010 = nice_ride_2010_df.groupby("Month Name")["Start date"].count()
month_count_2011 = nice_ride_2011_df.groupby("Month Name")["Start date"].count()
month_count_2012 = nice_ride_2012_df.groupby("Month Name")["Start date"].count()
month_count_2013 = nice_ride_2013_df.groupby("Month Name")["Start date"].count()
month_count_2014 = nice_ride_2014_df.groupby("Month Name")["Start date"].count()
month_count_2015 = nice_ride_2015_df.groupby("Month Name")["Start date"].count()
month_count_2016 = nice_ride_2016_df.groupby("Month Name")["Start date"].count()
month_count_2017 = nice_ride_2017_df.groupby("Month Name")["Start date"].count()
```

```python
Total_rides_2010 = nice_ride_2010_df["Start date"].count()
Total_rides_2011 = nice_ride_2011_df["Start date"].count()
Total_rides_2012 = nice_ride_2012_df["Start date"].count()
Total_rides_2013 = nice_ride_2013_df["Start date"].count()
Total_rides_2014 = nice_ride_2014_df["Start date"].count()
Total_rides_2015 = nice_ride_2015_df["Start date"].count()
Total_rides_2016 =nice_ride_2016_df["Start date"].count()
Total_rides_2017 = nice_ride_2017_df["Start date"].count()
Total_rides_2018 = nice_ride_2018_df["start_time"].count()
```

```python
total_rides = [['2010', Total_rides_2010], ['2011', Total_rides_2011], ['2012', Total_rides_2012], ['2013', Total_rides_2013], 
                 ['2014', Total_rides_2014],['2015', Total_rides_2015], ['2016', Total_rides_2016],
                ['2017', Total_rides_2017], ['2018', Total_rides_2018]]

# Create the pandas DataFrame
total_rides_df = pd.DataFrame(total_rides, columns = ['Years', 'Total Rides'])
total_rides_df
```

```python
plt.style.use("classic")

total_rides_df.plot(kind="bar", x="Years", width=1)

plt.title("Ridership (2010 - 2018)")
plt.ylabel("Number of Rides")
plt.xlabel("Years")
plt.grid(True)

plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))

plt.savefig("Nimo_YearRider.png", bbox_inches="tight")

plt.show()
plt.tight_layout()
```
