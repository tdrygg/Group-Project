### NiceRide Member vs Casual Account Type Comparison

**Question:**: How did subscriber ridership change over time? In this case ridership is measured by total number of rides for each account type.  How does changes in fee structure impact the account type? 

- From 2010-2017:  Member account type indicated users who had an account with NiceRide. For the same time period, Casual account type indicated users who walked up and bought a pass at the station.<br>


- 2018: Member account type indicated users who have a monthly or annual subscription membership with NiceRide. Casual account type indicated users who purchased a single ride or purchased a day pass.<br>


- From 2010 to 2014 Member account type members outpaced growth of casual riders.<br>


- From 2014-2016, membership growth remained flat and number of riders by account type did not differ greatly as casual rides increased.<br>


- 2016 saw a large dip in the number of casual riders. At this time, NiceRide offering a discount on single rides for users who create a membership and use the NiceRide app. This could help to explain the reduction in Casual account type riders at this point in time.<br>


- By 2018 there the number of member vs casual rides seemingly flipped as member riders steeply declined, and casual riders sharply rose. In 2018 NiceRide began offering a day pass option changing how member vs casual riders were classified.<br>

 
- The day pass popular choice among riders and shifted the distribution of Member and Casual account types.Incentivising usage with discounts or flexible pricing options is effective in driving users to a specific account type.<br>



```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#load data sets 

_2010_data = pd.read_csv("Project1CSV/Niceride 2010to2018/NR2010/Nice_Ride_trip_history_2010_season.csv", low_memory=False)
_2011_data = pd.read_csv("Project1CSV/Niceride 2010to2018/NR2011/Nice_Ride_trip_history_2011_season.csv", low_memory=False)
_2012_data = pd.read_csv("Project1CSV/Niceride 2010to2018/NR2012/Nice_Ride_trip_history_2012_season.csv", low_memory=False)
_2013_data = pd.read_csv("Project1CSV/Niceride 2010to2018/NR2013/Nice_Ride_trip_history_2013_season.csv", low_memory=False)
_2014_data = pd.read_csv("Project1CSV/Niceride 2010to2018/NR2014/Nice_Ride_trip_history_2014_season.csv", low_memory=False)
_2015_data = pd.read_csv("Project1CSV/Niceride 2010to2018/NR2015/Nice_Ride_trip_history_2015_season.csv", low_memory=False)
_2016_data = pd.read_csv("Project1CSV/Niceride 2010to2018/NR2016/Nice_Ride_trip_history_2016_season.csv", low_memory=False)
_2017_data = pd.read_csv("Project1CSV/Niceride 2010to2018/NR2017/Nice_Ride_trip_history_2017_season.csv", low_memory=False)
_2018_data = pd.read_csv("Project1CSV/Niceride 2010to2018/NR2018/data_2018.csv", low_memory=False)


#Drop rows with blank values (did not for 2018 due due dockless bike ride record had blank values for end time and station)

_2010_data = _2010_data.dropna(how="any")
_2011_data = _2011_data.dropna(how="any")
_2012_data = _2012_data.dropna(how="any")
_2013_data = _2013_data.dropna(how="any")
_2014_data = _2014_data.dropna(how="any")
_2015_data = _2015_data.dropna(how="any")
_2016_data = _2016_data.dropna(how="any")
_2017_data = _2017_data.dropna(how="any")

```

```python
#generate plotting variables and statistcs 

#calculate total rides for each year

total_2010 = len(_2010_data)
total_2011 = len(_2011_data)
total_2012 = len(_2012_data)
total_2013 = len(_2013_data)
total_2014 = len(_2014_data)
total_2015 = len(_2015_data)
total_2016 = len(_2016_data)
total_2017 = len(_2017_data)
total_2018 = len(_2018_data)

#calculate total member and casual rides in each year (ignoring 'Unknown'
#account type as it only accounts for 1% and less of total rides)

mem2010 = _2010_data.loc[_2010_data["Account type"] == "Member"]["Start date"].count()
mem2011 = _2011_data.loc[_2011_data["Account type"] == "Member"]["Start date"].count()
mem2012 = _2012_data.loc[_2012_data["Account type"] == "Member"]["Start date"].count()
mem2013 = _2013_data.loc[_2013_data["Account type"] == "Member"]["Start date"].count()
mem2014 = _2014_data.loc[_2014_data["Account type"] == "Member"]["Start date"].count()
mem2015 = _2015_data.loc[_2015_data["Account type"] == "Member"]["Start date"].count()
mem2016 = _2016_data.loc[_2016_data["Account type"] == "Member"]["Start date"].count()
mem2017 = _2017_data.loc[_2017_data["Account type"] == "Member"]["Start date"].count()
mem2018 = _2018_data.loc[_2018_data["usertype"] == "Member"]["start_time"].count()

cas2010 = _2010_data.loc[_2010_data["Account type"] == "Casual"]["Start date"].count()
cas2011 = _2011_data.loc[_2011_data["Account type"] == "Casual"]["Start date"].count()
cas2012 = _2012_data.loc[_2012_data["Account type"] == "Casual"]["Start date"].count()
cas2013 = _2013_data.loc[_2013_data["Account type"] == "Casual"]["Start date"].count()
cas2014 = _2014_data.loc[_2014_data["Account type"] == "Casual"]["Start date"].count()
cas2015 = _2015_data.loc[_2015_data["Account type"] == "Casual"]["Start date"].count()
cas2016 = _2016_data.loc[_2016_data["Account type"] == "Casual"]["Start date"].count()
cas2017 = _2017_data.loc[_2017_data["Account type"] == "Casual"]["Start date"].count()
cas2018 = _2018_data.loc[_2018_data["usertype"] == "Casual"]["start_time"].count()

#calculate percentage of rides taken by member user per year

mem_pct_2010 = (mem2010/total_2010) 
mem_pct_2011 = (mem2011/total_2011) 
mem_pct_2012 = (mem2012/total_2012) 
mem_pct_2013 = (mem2013/total_2013) 
mem_pct_2014 = (mem2014/total_2014) 
mem_pct_2015 = (mem2015/total_2015) 
mem_pct_2016 = (mem2016/total_2016) 
mem_pct_2017 = (mem2017/total_2017) 
mem_pct_2018 = (mem2018/total_2018) 

#calculate percentage of rides taken by casual user per year
cas_pct_2010 = (cas2010/total_2010) 
cas_pct_2011 = (cas2011/total_2011) 
cas_pct_2012 = (cas2012/total_2012) 
cas_pct_2013 = (cas2013/total_2013) 
cas_pct_2014 = (cas2014/total_2014) 
cas_pct_2015 = (cas2015/total_2015) 
cas_pct_2016 = (cas2016/total_2016) 
cas_pct_2017 = (cas2017/total_2017) 
cas_pct_2018 = (cas2018/total_2018) 


    
```

```python
#Read in as dataframe

acct_type_df = pd.DataFrame({"Year" : ["2010", "2011","2012", "2013", "2014", "2015", "2015", "2017", "2018"],
                            "Member Total Rides" : [mem2010, mem2011, mem2012, mem2013, mem2014, mem2015, mem2016,
                                                   mem2017, mem2018],
                            "Member Ride Pct" : [mem_pct_2010, mem_pct_2011, mem_pct_2012, mem_pct_2013, mem_pct_2014,
                                                 mem_pct_2015, mem_pct_2016, mem_pct_2017, mem_pct_2018],
                            "Casual Total Rides" : [cas2010, cas2011, cas2012, cas2013, cas2014, cas2015, cas2016,
                                                   cas2017, cas2018],
                            "Casual Ride Pct" : [cas_pct_2010, cas_pct_2011, cas_pct_2012, cas_pct_2013, cas_pct_2014,
                                                cas_pct_2015, cas_pct_2016, cas_pct_2017, cas_pct_2018]})

acct_type_df2 = acct_type_df.set_index("Year")

acct_type_df2["Member Ride Pct"] = acct_type_df2["Member Ride Pct"].map("{:.2%}".format)
acct_type_df2["Casual Ride Pct"] = acct_type_df2["Casual Ride Pct"].map("{:.2%}".format)

acct_type_df2
```

```python
#Plot total member and caual rides per year

x_axis = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]

yearly_member_count = [mem2010, mem2011, mem2012, mem2013, mem2014, mem2015, mem2016, mem2017, mem2018]

yearly_casual_count = [cas2010, cas2011, cas2012, cas2013, cas2014, cas2015, cas2016, cas2017, cas2018]

#using FiveThirtyEight style sheet
plt.style.use('fivethirtyeight')

#plot data together
plt.plot(x_axis, yearly_member_count, color="coral", marker="o", linewidth=1, label="Member")
plt.plot(x_axis, yearly_casual_count, color="teal", marker = "o", linewidth=1, linestyle="-", label="Casual")

#format plot
plt.title("Total Rides By Account Type")
plt.xlabel("Year")
plt.ylabel("Number of Rides")

plt.grid(color="white")

plt.legend(title="Account Type", title_fontsize=16, fancybox=True, frameon=True, 
           shadow=True, facecolor="white", fontsize=14)

#text_label= "*Day pass  pricing tier was added in 2018. Users who"
#text_label2="purchased this were classified as a casual account type"
#plt.text(2018,100000, text_label, fontsize=13, ha='left')
#plt.text(2018,85000, text_label2, fontsize=13, ha='left')

plt.tight_layout()

plt.savefig("total_rides_by_acct_type")
```
